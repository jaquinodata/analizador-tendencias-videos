Vamos a comprobar si los registros de los ficheros json contienen valores repetidos para el campo id de categoría.

```python
k = 0
for dataframe in json_dict.values():
    lista_ids = []
    k += 1
    for i in dataframe['items']:
        lista_ids.append(i['id'])
    contador = Counter(lista_ids)
    repetidos = [item for item, count in contador.items() if count > 1]
    if repetidos:
        print(f'Valores repetidos en el fichero {k}: {repetidos}')
    else:
        print('No hay valores repetidos en el fichero ' + str(k))
```

## Unificación de los datos

A continuación, vamos a recoger la información de categorías del fichero json de cada país y a añadirla al fichero csv de dicho país.

```python
csv_copia = csv.copy()
```

```python
csv_copia['CA']['category_id'].head()
```

Empezamos con los datos de CA (Canadá)

```python
country = 'CA'
lista_titulos = []
# recorre todas las filas del fichero csv
for fila in range(0, csv_copia[country].shape[0]):
    id = csv_copia[country].loc[fila]['category_id']
    titulo = ''
    # busca el id de categoría en el fichero json
    for j in json_dict[country]['items']:
        if int(j['id']) == id:
            # categoría encontrada en el fichero json
            titulo = j['snippet']['title']
            break

    if titulo == '':
        titulo = 'categoría no encontrada en el fichero json'
    # guarda el nombre de la categoría
    lista_titulos.append(titulo)

# añadimos la lista de nombres de categorías al dataframe
csv_copia[country]['category'] = lista_titulos

# añadimos el código de país al dataframe
csv_copia[country]['country'] = country
display(csv_copia[country][['category_id', 'category', 'country']].head(10))
```

Repetimos el proceso para todos los dataframes

```python
# guarda en dict_cats todas las categorías de cada país
# dict_cats: { cod_pais: {id_cat: nombre_cat}, ... }
country_list = ['CA', 'DE', 'FR', 'GB', 'IN', 'JP', 'KR', 'MX', 'RU', 'US']
dict_cats = {}
for country in country_list:
    dict = {}
    for j in json_dict[country]['items']:
        if j['id'] not in dict:
            dict[j['id']] = j['snippet']['title']
    dict_cats[country] = dict

print(dict_cats)
```

Vamos a revisar los ids de categoría de todos los países

```python
for i in iter(dict_cats):
    ids_cat = []
    for j in iter(dict_cats[i]):
        ids_cat.append(j)
    print(i, ':', ids_cat)
```

Todos los países tienen los mismos ids de categoría, salvo US que tiene además el 29.

```python
# añade en csv_category una columna con el título de la categoría y otra con el código de país
csv_category = csv.copy()

for country in country_list:
    lista_titulos = []
    # recorre todas las filas del fichero csv
    for fila in range(0, csv_category[country].shape[0]):
        id = csv_category[country].loc[fila]['category_id']
        titulo = ''
        # busca el id de categoría en el fichero json
        for j in json_dict[country]['items']:
            if int(j['id']) == id:
                # categoría encontrada en el fichero json
                titulo = j['snippet']['title']
                break

        if titulo == '':
            titulo = 'categoría no encontrada en el fichero json'
        # guarda el nombre de la categoría
        lista_titulos.append(titulo)

    # añadimos la lista de nombres de categorías al dataframe
    csv_category[country]['category'] = lista_titulos

    # añadimos el código de país al dataframe
    csv_category[country]['country'] = country
    
    display(csv_category[country][['category_id', 'category', 'country']].head(2))
```

A continuación, vamos a crear un único dataframe que combine los dataframes de todos los países.

```python
for country in country_list:
    print(csv_category[country].shape)
```

```python
# recorremos los países, concatenando los csv uno por uno al primero
kont = 1
for country in country_list:
    if kont == 1:
        df = csv_category[country].copy()
    else:
        df = pd.concat([df, csv_category[country]], ignore_index=True)
    kont += 1
df.shape
```
