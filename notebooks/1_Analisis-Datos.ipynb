Vamos a comprobar si los registros de los ficheros json contienen valores repetidos para el campo id de categorÃ­a.

```python
k = 0
for dataframe in json_dict.values():
    lista_ids = []
    k += 1
    for i in dataframe['items']:
        lista_ids.append(i['id'])
    contador = Counter(lista_ids)
    repetidos = [item for item, count in contador.items() if count > 1]
    if repetidos:
        print(f'Valores repetidos en el fichero {k}: {repetidos}')
    else:
        print('No hay valores repetidos en el fichero ' + str(k))
```

## UnificaciÃ³n de los datos

A continuaciÃ³n, vamos a recoger la informaciÃ³n de categorÃ­as del fichero json de cada paÃ­s y a aÃ±adirla al fichero csv de dicho paÃ­s.

```python
csv_copia = csv.copy()
```

```python
csv_copia['CA']['category_id'].head()
```

Empezamos con los datos de CA (CanadÃ¡)

```python
country = 'CA'
lista_titulos = []
# recorre todas las filas del fichero csv
for fila in range(0, csv_copia[country].shape[0]):
    id = csv_copia[country].loc[fila]['category_id']
    titulo = ''
    # busca el id de categorÃ­a en el fichero json
    for j in json_dict[country]['items']:
        if int(j['id']) == id:
            # categorÃ­a encontrada en el fichero json
            titulo = j['snippet']['title']
            break

    if titulo == '':
        titulo = 'categorÃ­a no encontrada en el fichero json'
    # guarda el nombre de la categorÃ­a
    lista_titulos.append(titulo)

# aÃ±adimos la lista de nombres de categorÃ­as al dataframe
csv_copia[country]['category'] = lista_titulos

# aÃ±adimos el cÃ³digo de paÃ­s al dataframe
csv_copia[country]['country'] = country
display(csv_copia[country][['category_id', 'category', 'country']].head(10))
```

Repetimos el proceso para todos los dataframes

```python
# guarda en dict_cats todas las categorÃ­as de cada paÃ­s
# dict_cats: { cod_pais: {id_cat: nombre_cat}, ... }
country_list = ['CA', 'DE', 'FR', 'GB', 'IN', 'JP', 'KR', 'MX', 'RU', 'US']
dict_cats = {}
for country in country_list:
    dict = {}
    for j in json_dict[country]['items']:
        if j['id'] not in dict:
            dict[j['id']] = j['snippet']['title']
    dict_cats[country] = dict

print(dict_cats)
```

Vamos a revisar los ids de categorÃ­a de todos los paÃ­ses

```python
for i in iter(dict_cats):
    ids_cat = []
    for j in iter(dict_cats[i]):
        ids_cat.append(j)
    print(i, ':', ids_cat)
```

Todos los paÃ­ses tienen los mismos ids de categorÃ­a, salvo US que tiene ademÃ¡s el 29.

```python
# aÃ±ade en csv_category una columna con el tÃ­tulo de la categorÃ­a y otra con el cÃ³digo de paÃ­s
csv_category = csv.copy()

for country in country_list:
    lista_titulos = []
    # recorre todas las filas del fichero csv
    for fila in range(0, csv_category[country].shape[0]):
        id = csv_category[country].loc[fila]['category_id']
        titulo = ''
        # busca el id de categorÃ­a en el fichero json
        for j in json_dict[country]['items']:
            if int(j['id']) == id:
                # categorÃ­a encontrada en el fichero json
                titulo = j['snippet']['title']
                break

        if titulo == '':
            titulo = 'categorÃ­a no encontrada en el fichero json'
        # guarda el nombre de la categorÃ­a
        lista_titulos.append(titulo)

    # aÃ±adimos la lista de nombres de categorÃ­as al dataframe
    csv_category[country]['category'] = lista_titulos

    # aÃ±adimos el cÃ³digo de paÃ­s al dataframe
    csv_category[country]['country'] = country
    
    display(csv_category[country][['category_id', 'category', 'country']].head(2))
```

A continuaciÃ³n, vamos a crear un Ãºnico dataframe que combine los dataframes de todos los paÃ­ses.

```python
for country in country_list:
    print(csv_category[country].shape)
```

```python
# recorremos los paÃ­ses, concatenando los csv uno por uno al primero
kont = 1
for country in country_list:
    if kont == 1:
        df = csv_category[country].copy()
    else:
        df = pd.concat([df, csv_category[country]], ignore_index=True)
    kont += 1
df.shape
```
